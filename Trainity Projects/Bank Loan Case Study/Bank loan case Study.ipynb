3.Indentifying Missing Data
i)In df1(application_data)
ii)In df2(previous_application)

i)Missing Data in application_data

```python
null_percentage=round(df1.isna().sum().sort_values(ascending=False)/len(df1)*100,2)
null_percentage
```

```python
col_to_drop = null_percentage[null_percentage>45].keys()

output_df1 = df1.drop(col_to_drop, axis=1)
```

```python
output_df1.head()
```

```python
round(output_df1.isna().sum().sort_values(ascending=False)/len(df1)*100,2)
```

```python
output_df1.columns.nunique(),df1.columns.nunique(),col_to_drop.nunique()
```

```python
output_df1.shape
```

```python
output_df1.OCCUPATION_TYPE.value_counts()
```

```python
# Dealing with missing value
output_df1.OCCUPATION_TYPE.replace(np.NaN,"Other",inplace=True)
output_df1.NAME_TYPE_SUITE.replace(np.NaN,"Other_C",inplace=True)
```

```python
output_df1.NAME_TYPE_SUITE.value_counts()
```

```python
null_percentage=round(df2.isna().sum().sort_values(ascending=False)/len(df2)*100,2)
null_percentage
```
